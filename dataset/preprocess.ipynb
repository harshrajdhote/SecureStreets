```python
import pandas as pd 
import csv
df = pd.read_csv("data_modified.csv")
primary_type = pd.read_csv("prime_crime_types.csv")
# print("Community_AREA    PRIMARY_TYPE     COUNT")
with open('crime_type_counts.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Community_AREA", "PRIMARY_TYPE", "COUNT"])

    for index, row in primary_type.iterrows():
        print(row['PRIMARY TYPE'])
        writer.writerow([row['PRIMARY TYPE']])
        # print("Community_AREA    PRIMARY_TYPE     COUNT")
        for i in range (1,78):    
            com_area_crime = df[(df['Community Area'] == i) & (df['Primary Type'] == row['PRIMARY TYPE']) ]
            writer.writerow([i,row['PRIMARY TYPE'],com_area_crime.shape[0]])
```
